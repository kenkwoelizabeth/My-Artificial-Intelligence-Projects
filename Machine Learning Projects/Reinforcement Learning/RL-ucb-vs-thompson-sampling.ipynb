### Thompson Sampling Vs Upper Confidence Bound:

Keeping in mind that we all have basic intution behind the both reinforcement learning algorithms, today I want to showcase
both algorithms one by one on the same dataset and I will reveal the difference behind them at the end.

Here, let's take a hypothetical situation wherein we have been hired as a datascientist for XYZ firm which wants to run an
advertisement campaign for it's product but not sure which one would entice more customers and we have to explore the data
in such a way that it would maximize the return on investment while we explore the data.

Our purpose is to find the ideal trade off between exploring and expoiting.

### Thompson Sampling:

We will start with Thompson algorithm and then at the end we will comare the results of both the algorithms.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
dataset = pd.read_csv('../input/ten-ads-optimization/Ads_CTR_Optimisation.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
observations = 10000
Ads = 10

number_of_reward_1 = [0]*Ads
number_of_reward_0 = [0]*Ads

selected_ad = []
total_reward = 0

for n in range(observations):
    index = 0
    max_random = 0
    for i in range(Ads):
        random_beta = random.betavariate(number_of_reward_1[i]+1, number_of_reward_0[i]+1)
        if random_beta > max_random:
            max_random = random_beta
            index = i

    selected_ad.append(index)
    reward = dataset.values[n,index]
    if reward == 1:
        number_of_reward_1[index] += 1
    else:
        number_of_reward_0[index] += 1
    total_reward = total_reward + reward
```

```python
print(total_reward)
print(number_of_reward_0)
print(number_of_reward_1)
```

```python
plt.style.use('dark_background')
```

```python
plt.hist(selected_ad,edgecolor='black')
plt.title('Histogram of Ads selected')
plt.xlabel('Advertisements')
plt.ylabel('Number of time Ad was selected')
plt.show()
```

```python
plt.hist(selected_ad,edgecolor='black')
plt.title('Histogram of Ad selection')
plt.xlabel('Advertisements')
plt.ylabel('Number of times Ad selected')
plt.show()
```

### Upper Confidence Bound

```python
dataset_ucb = pd.read_csv('../input/ten-ads-optimization/Ads_CTR_Optimisation.csv')
```

```python
observations_ucb = 10000
Ads_ucb = 10

number_of_selection_of_each_ad = [0]*Ads_ucb
sum_of_reward_of_each_ad = [0]*Ads_ucb

selected_ad_ucb = []
total_reward_ucb = 0

for n in range(observations_ucb):
    index_ucb = 0
    max_upper_bound = 0
    for i in range(Ads_ucb):
        if number_of_selection_of_each_ad[i] > 0:
            average_reward = sum_of_reward_of_each_ad[i]/number_of_selection_of_each_ad[i]
            delta = np.sqrt(3/2 * np.log(n + 1)/ number_of_selection_of_each_ad[i])
            upper_bound = average_reward + delta

        else:
            upper_bound = 1e400

        if upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            index_ucb = i

    selected_ad_ucb.append(index_ucb)
    number_of_selection_of_each_ad[index_ucb] += 1
    reward_ucb = dataset_ucb.values[n,index_ucb]
    sum_of_reward_of_each_ad[index_ucb] = sum_of_reward_of_each_ad[index_ucb] + reward_ucb
    total_reward_ucb = total_reward_ucb + reward_ucb
```

```python
print(number_of_selection_of_each_ad)
print(sum_of_reward_of_each_ad)
```

```python
plt.hist(selected_ad_ucb,edgecolor='black')
plt.title('Histogram of Ad selection')
plt.xlabel('Advertisements')
plt.ylabel('Number of time Ad selected')
plt.show()
```

### Conclusion:

We can clearly see that eventhough the number of sampling decreases Thompson algorithm is able to identify the Ad which can earn optimal return on investment.

The other draw back of UCB is it requires to update the information at each round which may not be possible in each case. Like in this case suppose we have a threshold that after 500 clicks of Ad we dump the data from website, in this case Thompson sampling can be useful because it accomodate for delayed feedback.
